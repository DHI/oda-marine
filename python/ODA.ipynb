```python
from zeep import Client
from datetime import datetime
import pandas as pd

client = Client('http://odahavdata.oda.dk/ODAHavData.asmx?WSDL')
```

# Possible subjects

```python
subjects = client.service.dataemner()

data =list(map(lambda x: {'id' : x.id, 'value': x.value}, subjects))

pd.DataFrame(data)
```

# Getting data by Station name

```python

rawdata = client.service.getDataByStation(startDT=datetime(2018,1,1),
                            slutDT=datetime(2018,2,1),
                            dataemneId=10,
                            parameterId=251,
                            stationName= "VSJ20925")

data =list(map(lambda x: {'station': x.stationNavn,
                          'lon': x.lon,
                          'lat': x.lat,
                          'date': x.dato,
                          'parameter_id'  :x.parameterId,
                          'unit' : x.enhed,
                          'depth' : x.dybde,
                          'value' : x.vaerdi,
                          'qa_level' : x.QAniveau,
                          'species_id': x.ArtId}, 
                           rawdata))

df =pd.DataFrame(data)
df

#df[['date','value']]

```
